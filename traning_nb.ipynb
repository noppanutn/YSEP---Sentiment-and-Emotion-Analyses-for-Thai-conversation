```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import GaussianNB
import tensorflow
import deepcut
from pythainlp.corpus import stopwords
import re
import matplotlib.pyplot as plt
```

```python
myData=[]
i=1
filename = "./dataset/forCLFnohash1label1000.csv"
```

```python
def tokenizer(tweet):
    #global i
    #i = i+1
    #if i%100 == 0: print("%d from %d" (i,len(myData)))
    pattern = re.compile(r"[^\u0E00-\u0E4C]|^'|'$|''|\u0E46|\u0E3F")
    char_to_remove = re.findall(pattern, tweet)
    list_with_char_removed = [char for char in tweet if not char in char_to_remove]
    result_string = ''.join(list_with_char_removed)
    return deepcut.tokenize(result_string)
```

```python
import csv
with open(filename, 'r', encoding='utf8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        myData.append(row["tweet"])
```

```python
df = pd.read_csv(filename)
y = np.array(df['label'])
print(y)
```

```python
v = TfidfVectorizer(tokenizer=tokenizer, stop_words=stopwords.words('thai'))
train_data_feature = v.fit_transform(myData).toarray()
print(train_data_feature)
#tokenizer("ไม่รู้เหมือนกัน")
```

```python
vocab = v.get_feature_names()
#print(vocab)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
X, y = train_data_feature, y
OneVsRestClassifier(LinearSVC(random_state=0)).fit(X,y).predict(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data_feature, y, test_size=0.2, random_state=0)
clf = OneVsRestClassifier(LinearSVC(random_state=0)).fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
class MeanEmbeddingVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        # if a text is empty we should return a vector of zeros
        # with the same dimensionality as all the other vectors
        self.dim = len(list(word2vec.values()).next())

    def fit(self, X, y):
        return self

    def transform(self, X):
        return np.array([
            np.mean([self.word2vec[w] for w in words if w in self.word2vec]
                    or [np.zeros(self.dim)], axis=0)
            for words in X
        ])
    
```

```python
class TfidfEmbeddingVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        self.word2weight = None
        self.dim = len(word2vec.values().next())

    def fit(self, X, y):
        tfidf = TfidfVectorizer(analyzer=lambda x: x)
        tfidf.fit(X)
        # if a word was never seen - it must be at least as infrequent
        # as any of the known words - so the default idf is the max of 
        # known idf's
        max_idf = max(tfidf.idf_)
        self.word2weight = defaultdict(
            lambda: max_idf,
            [(w, tfidf.idf_[i]) for w, i in tfidf.vocabulary_.items()])

        return self

    def transform(self, X):
        return np.array([
                np.mean([self.word2vec[w] * self.word2weight[w]
                         for w in words if w in self.word2vec] or
                        [np.zeros(self.dim)], axis=0)
                for words in X
            ])
```

```python
def vectorize_feature(sentence):
    s_vector = np.zeros(100)
    for word in sentence:
        if word in model.wv.vocab:
            s_vector = s_vector + model.wv.__getitem__(word)
    return s_vector
print(lw[1])
vectorize_feature(lw[1])
```

```python
def extract_feature(list_of_sentences):
    f_vector = []
    for sentence in list_of_sentences:
        f_vector.append(vectorize_feature(sentence))
    return f_vector
```
